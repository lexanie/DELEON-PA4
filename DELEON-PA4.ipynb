```python
# import matplot

import matplotlib.pyplot as plt
```

```python
board2.plot.hist(x='Name', y=['Math', 'GEAS', 'Electronics', 'Communication'])
```

```python
combined_avg = board2.groupby(['Track', 'Gender', 'Hometown'])['Average'].mean().unstack()
print("\nCombined Average Grade by Track, Gender, and Hometown:")
print(combined_avg)

fig= plt.subplots(1, 1, figsize=(5,6))

# Average Grade by Track
track_avg = board2.groupby('Track')['Average'].mean()
track_avg.plot(kind='bar', color='#F0A8D0', title='Average Grade by Track', ylabel='Average Grade')
```

```python
# Average Grade by Gender
gender_avg = board2.groupby('Gender')['Average'].mean()
gender_avg.plot(kind='bar', color='#F7B5CA', title='Average Grade by Gender', ylabel='Average Grade')
```

```python
# Average Grade by Hometown
hometown_avg = board2.groupby('Hometown')['Average'].mean()
hometown_avg.plot(kind='bar', color='#FFC6C6', title='Average Grade by Hometown', ylabel='Average Grade')
```
